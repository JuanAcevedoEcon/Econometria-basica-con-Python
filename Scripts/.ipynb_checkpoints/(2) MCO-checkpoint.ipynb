# Condicionales

**if - else**  


```python
edad = int(input("¿Cuál es su edad?"))

if edad < 18:
    print("Usted es menor de edad")
else:
    print("Usted es mayor de edad")
```

```python
def valor_absoluto(x):
    if x >= 0:
        valor = x
    else:
        valor = -x
    return valor

x = float(input("Ingrese un número real x: "))
valor_abs = valor_absoluto(x)
print("El valor absoluto de " + str(x), end = " ")
print("es " + str(valor_abs))
```

**enlazado (elif)**  


```python
# Edad en montaña rusa
edad = int(input("¿Cuál es su edad?"))


if edad < 18:
    print("Usted es menor de edad y no puede entrar")
elif edad > 80:
    print("Usted es mayor de edad y no puede entrar por posibles problemas en el corazón")
else:
    print("Usted es mayor de edad y está en edad para entrar")
```

Una tienda tiene las siguientes promociones

>Si un cliente lleva más de 5 productos del mismo tipo le realizan un descuento del 5%. Si lleva más de 10 productos del
mismo tipo le realizan un descuento del 10%. Si lleva más de 20
productos del mismo tipo le realizan un descuento del 20%.
Construya un programa que dado el número de productos y el
precio de cada producto determine el valor a pagar por el cliente.


```python
def pago_final(n, precio):
    if n <= 5:
        valor = n * precio
    elif 5 < n <= 10:
        valor = n * precio * 0.95
    elif 10 < n <= 20:
        valor = n * precio * 0.90
    else:
        valor = n * precio * 0.80
    return valor
```

```python
pago_final(8,10000)
```

## Estructuras cíclicas


**while (mientras)** 

Se ejecuta mientras la expresión booleana se cumpla (True). 



```python
Edad = 0

while(Edad < 18):
    print("Usted cree en Papa Noel")
    Edad += 3
```

```python
i = 0
while(i <= 6):
    print(i)
    i = i + 1 # i += 1
```

```python
i = 2
j = 25 
while i < j: 
    print(i, j, sep = ", ")  
    i *= 2
    j += 10 
print("the end.") 
print(i, j, sep = ", ") # valores finales de i y j
```

```python
# Uso de banderas #

sumafinal = 0
bandera = True
print("Bienvenido a la sumatoria de Python")
print("Nota: cuando la persona ingrese 0 se finaliza la suma")

while(bandera):
    x = float(input("Inserte un número para sumar"))
    if x == 0:
        bandera = False
    else:
        sumafinal += x
        
print(sumafinal)
```

**for (para)**  
Para iterar sobre los elementos de una colección de elementos.

```python
frutas = ["Tomate de árbol", "Maracuyá ", "Guayaba"]
for i in frutas: 
    print(i)
```

```python
# Range(C,F,I) Cominzo, Final (no se incluye), Incremento/decremento

for f in range(2,13,2):
    print(f)
```

```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
datos= pd.read_csv("C:/Users/57317/Desktop/Universidad Nacional/Programación/Python/Curso Econometría básica con Python/Bases de datos/GujaratiPorter71.txt", delimiter="\t")
```

```python
datos.head()
```

```python
fig, ax = plt.subplots() # Se ponen dos nombres hay una figura y dentro un ax (otra)
ax.scatter(datos["PGNP"],datos["CM"])
ax.set_xlabel("PNB per cápita")
ax.set_ylabel("Mortalidad infantil")
ax.set_title("Relación", fontsize = 15, color="black")
plt.show()
```

```python
Modelo=sm.OLS(datos["CM"], datos["PGNP"]) #Primero y después x
Resultados = Modelo.fit()
print(Resultados.summary())
```

```python
Modelo2=sm.OLS(datos["CM"],sm.add_constant(datos["PGNP"])) #Primero y después x
Resultados2 = Modelo2.fit()
print(Resultados2.summary())
```

```python
Valores_estimados = Resultados2.predict()
```

```python
fig, ax = plt.subplots()
ax.scatter(datos["PGNP"],datos["CM"], color="blue")
ax.scatter(datos["PGNP"],Valores_estimados, color="red")
ax.set_xlabel("PNB per capita")
ax.set_ylabel("Mortalidad Infantil")
ax.set_title("Relación", fontsize = 15, color="black")
plt.show()
```
