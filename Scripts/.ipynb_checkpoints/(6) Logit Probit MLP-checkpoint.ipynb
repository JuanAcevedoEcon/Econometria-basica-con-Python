```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.stats.api as sms
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_table("http://www.econometrics.com/comdata/gujarati/data_15.7.shd", header = None,
                    delim_whitespace = True,
                    names = ["Fumador","Edad","Escolaridad","Ingresos","Pcigs79"]) # Cargar de internet
```

```python
data.head()
```

```python
grafico = sns.pairplot(data)
plt.show()
```

```python
data.describe()
```

```python
fig, ax = plt.subplots()
ax.hist(data["Fumador"])
plt.show()
```

```python
Y = data["Fumador"]
X = data[["Edad","Escolaridad","Ingresos","Pcigs79"]]
```

```python
MLP = sm.OLS(Y, sm.add_constant(X))
ResultadosMLP = MLP.fit()
print(ResultadosMLP.summary())
```

```python
MLPMargEff = ResultadosMLP.params 
```

```python
MLPMargEff
```

```python
PrediccionMLP = ResultadosMLP.predict()
PrediccionMLP = pd.DataFrame(PrediccionMLP, columns = ["Prediccion MLP"])
```

```python
PrediccionMLP.describe()
```

```python
ModeloLogit = sm.Logit(Y, sm.add_constant(X))
ResultadosLogit = ModeloLogit.fit()
print(ResultadosLogit.summary())
```

```python
LogitMargEff = ResultadosLogit.get_margeff() # Efectos marginales PROMEDIO
print(LogitMargEff.summary())
```

```python
LogitMargEff = ResultadosLogit.get_margeff(at = "mean")
print(LogitMargEff.summary())
```

**Modelo Probit**

```python
ModeloProbit = sm.Probit(Y,sm.add_constant(X))
ResultadosProbit = ModeloProbit.fit()
print(ResultadosProbit.summary())
```

```python
ProbitMarEff = ResultadosProbit.get_margeff()
print(ProbitMarEff.summary())
```

```python
ProbitMarEff = ResultadosProbit.get_margeff(at = "mean")
print(ProbitMarEff.summary())
```

```python
from statsmodels.iolib.summary2 import summary_col
```

```python
Tabla = summary_col(results=[ResultadosMLP,ResultadosLogit,ResultadosProbit],
                   model_names = ["MLP","Logit","Probit"], stars= True)
print(Tabla)
```

```python
Coeficientes = ["Edad","Escolaridad","Ingresos","Pcigs79"]
EfectosLogit = pd.DataFrame(LogitMargEff.margeff,
                           index= Coeficientes,
                           columns = ["Logit"])
EfectosProbit = pd.DataFrame(ProbitMarEff.margeff,
                           index= Coeficientes,
                           columns = ["Probit"])
EfectosMLP = pd.DataFrame(MLPMargEff[1:5],
                         columns = ["MLP"])
```

```python
pd.concat([EfectosLogit, EfectosProbit, EfectosMLP], axis = 1)
```
